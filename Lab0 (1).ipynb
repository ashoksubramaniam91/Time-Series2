```python
newVariable = ['q1', 'q2', 'q3', 'q4','q1', 'q2', 'q3', 'q4','q1', 'q2', 'q3', 'q4',
               'q1', 'q2', 'q3', 'q4', 'q1', 'q2', 'q3', 'q4']
```

```python
df.head()
```

```python
dfComplete = pd.get_dummies(df, drop_first=True)
```

```python
dfComplete.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
x = dfComplete.drop('Sales1000s', axis=1)
```

```python
y = dfComplete[['Sales1000s']]
```

```python
model.fit(x, y)
```

```python
predictions = model.predict(x)
```

```python
from matplotlib import pyplot
```

```python
pyplot.plot(y)
pyplot.plot(predictions, color='red')
pyplot.show()
```
